```python
# Re-do the plot from the previous cell as a horizontal bar plot

df.iloc[:10].plot(kind='barh', 
                  x='Education.Major', 
                  y= 'Salaries.Mean')
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)
barcolors = []
for i in df[:10].index:
    if df.loc[i,'Education.Major'] in ['Economics']:
        barcolors.append('green')
    else:
        barcolors.append('blue')
print(barcolors)
```

```python
df[:10].plot(kind='barh',
             x='Education.Major',
             y='Salaries.Mean',   # <------ this line has to be included
             figsize=(13,8),
             color=barcolors)
```
